# Get tweets that may mention extremist events

```python
import tweepy
import pandas as pd
```

```python
# Load authentication
keys = pd.read_csv('/storage2/mamille3/tweepy_oauth_academic.csv', index_col='name').to_dict()['key']
```

```python
# auth = tweepy.OAuthHandler(keys["consumer_key"], keys['consumer_secret'])
# auth.set_access_token(keys['access_token'], keys['access_secret'])
```

```python
client = tweepy.Client(keys['bearer_token'])
```

```python
from datetime import datetime

tweet_fields = [
    'id', 'created_at', 'text', 'author_id', 'conversation_id', 'entities', 'public_metrics', 'geo', 'lang', 'referenced_tweets'
]
place_fields = [
    'full_name', 'id', 'contained_within'
]

# For fire at an abortion clinic in Toledo https://nbc24.com/news/local/fire-at-west-toledo-abortion-clinic-ruled-arson
query = '"abortion clinic" (fire OR arson) toledo'
start_time = datetime(2021,10,31)
end_time = datetime(2021,11,14)

# For bottle thrown at Jewish youth walking back from school in Cleveland https://www.cleveland19.com/2021/10/08/university-heights-police-search-woman-responsible-anti-semitic-attack-against-teen/
# query = '("university heights" OR cleveland) attack (anti-semitic OR antisemitic)'
# start_time = datetime(2021,10,7)
# end_time = datetime(2021,10,21)

fetched = client.search_all_tweets(query, expansions='geo.place_id', place_fields=place_fields, tweet_fields=tweet_fields, start_time=start_time, end_time=end_time, max_results=20)
# type(fetched)
fetched
```

```python
orig_tweets = [f for f in fetched.data if not 'referenced_tweets' in f.data]
print(len(orig_tweets))

print("ORIGINAL TWEETS:")
for tweet in orig_tweets:
    print(tweet.id)
    print(tweet.text)
    print(tweet.entities)
    print()
```

```python
rt_with_replies = [f for f in fetched.data if 'referenced_tweets' in f.data and f.data['referenced_tweets'][0]['type']=='quoted']
print(len(rt_with_replies))

print("RETWEETS WITH REPLIES:")
for tweet in rt_with_replies:
    print(tweet.id)
    print(tweet.text)
    print(tweet.entities)
    print()
```

```python
rts = [f for f in fetched.data if 'referenced_tweets' in f.data]
print(len(rts))

print("RETWEETS")
for tweet in rts:
    print(tweet.id)
    print(tweet.text)
    print(tweet.referenced_tweets)
    print()
```
